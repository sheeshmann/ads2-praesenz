# C - Modelle aus der Registry nutzen

## Imports

```python
import mlflow.pyfunc
import pandas as pd
```

## Modell aus der Registry laden

```python
mlflow.set_tracking_uri("sqlite:///metadata/mlflow/mlruns.db")
```

```python
model_name = "taxi_fare_regressor"
model_version = 1
model = mlflow.pyfunc.load_model(model_uri=f"models:/{model_name}/{model_version}")
```

## Testdaten laden & Predictions erzeugen

```python
test_data = pd.read_parquet("data/test_data_taxi_fares.parquet")
```

```python
model.predict(test_data)
```
